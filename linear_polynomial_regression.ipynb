Zad. 1 Korzystając z narzędzi dostępnych w języku Python rozwiąż następujące zadanie
regresji liniowej

Podzadania:
a) Wygeneruj odpowiedni zbiór punktów
b) Zaproponuj model odpowiedni dla wygenerowanych danych
c) Przygotuj w języku Python kod do estymacji parametru(ów) dla modelu z pkt. b


```python
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

x, y = datasets.make_regression(n_samples= 200, n_features= 1, noise= 25, random_state=7) #generujemy dane
linreg = LinearRegression()                                                               #regresja liniowa
linreg.fit(x.reshape(-1,1), y.reshape(-1,1))                                              #estymator                             
                                                                                          #a = Y.T @ X_ext @ np.linalg.inv(X_ext.T @ X_ext)
a, b = linreg.coef_, linreg.intercept_                                                    #szacowana zmienna i wyraz wolny
print(f'a= {a[0][0]} b= {b[0]}')

y_pred = linreg.predict(x.reshape(-1,1))
plt.scatter(x,y, color='green', marker='.')
plt.plot(x, y_pred, color='black', linewidth=3)
plt.show()

```

Zad. 2 Korzystając z narzędzi dostępnych w języku Python rozwiąż następujące zadanie
regresji wielomianowej

Podzadania:
a) Wygeneruj odpowiedni zbiór punktów
b) Zaproponuj model odpowiedni dla wygenerowanych danych
c) Przygotuj w języku Python kod do estymacji parametru(ów) dla modelu z pkt. b

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression

polyreg=make_pipeline(PolynomialFeatures(2),LinearRegression())

polyreg2=make_pipeline(PolynomialFeatures(3),LinearRegression())

df = sns.load_dataset('mpg').dropna()
x = df['horsepower'].to_numpy()
y = df['mpg'].to_numpy()
polyreg.fit(x.reshape(-1,1),y.reshape(-1,1))
polyreg2.fit(x.reshape(-1,1),y.reshape(-1,1))
x_lin = np.linspace(x.min(),x.max(),300).reshape(-1,1)

plt.scatter(x,y,color= 'green', marker= '.')
plt.plot(x_lin,polyreg.predict(x_lin),color="black", linewidth=1)
plt.plot(x_lin,polyreg2.predict(x_lin),color="red", linewidth=1)
plt.show()
```
