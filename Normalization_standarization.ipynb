```python
from sklearn import preprocessing
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

Zad. 1 Dla przykładowego szeregu czasowego rozwiąż zadanie normalizacji korzystając z
dostępnych w języku Python narzędzi. Przygotuj odpowiedni kod w tym języku. Zilustruj
efekty normalizacji przykładowego szeregu czasowego. 

When you normalize data, you change the scale of the data. Data is commonly rescaled to fall between 0 and 1, because machine learning algorithms tend to perform better, or converge faster, when the different features are on a smaller scale.
Normalization also makes the training process less sensitive to the scale of the features, resulting in better coefficients after training.

Xnorm = (X-Xmin)/(Xmax-Xmin)


```python
x = np.linspace(1,10, 100)
y= x**2
s = np.random.randint(0, 20, (100))
y = y+s
df = pd.DataFrame(y)

figure, axis = plt.subplots(1, 2, figsize=(16, 6))
axis[0].plot(x, y)
y = y.reshape(-1, 1)
scaler = preprocessing.MinMaxScaler()
y1 = (y-min(y))/(max(y)-min(y))
y_scal = scaler.fit_transform(y)
axis[1].plot(x, y_scal, color='orange', label='Normalized with sklearn function')
axis[1].scatter(x, y1, color='green', label='Normalized', marker='.')
plt.legend()
plt.show()
```

Zad. 2 Wygeneruj szereg czasowy. Korzystając z narzędzi dostępnych w języku Python,
przygotuj kod który dokona standaryzacji wygenerowanego szeregu czasowego. Efekty
standaryzacji zilustruj odpowiednimi wykresami.

```python
x = np.linspace(1,10, 100)
y= x**2
s = np.random.randint(0, 20, (100))
y = y+s

df = pd.DataFrame(y)
x = np.linspace(1, 1000, 100)  
df1 = (df - df.mean()) / df.std()

figure, axis = plt.subplots(1, 2, figsize=(16, 6))
axis[0].plot(x, df)
axis[1].plot(x, df1, color='orange', label='Standarized')
plt.legend()
plt.show()
```

Zad. 3 Dla syntetycznego szeregu czasowego rozwiąż zadanie skalowania przy wykorzystaniu
dostępnych w języku Python narzędzi. Przygotuj zestaw wykresów, które zilustrują efekt
działania operacji skalowania dla wygenerowanego szeregu czasowego.

```python
x = np.linspace(1,10, 100)
y= x**2
s = np.random.randint(0, 20, (100))
y = y+s

df = pd.DataFrame(y)
x = np.linspace(1, 1000, 100) 

figure, axis = plt.subplots(1, 2, figsize=(16, 6))
axis[0].plot(x, y)
y = y.reshape(-1, 1)
scaler = preprocessing.MinMaxScaler((10,20))
y_scal = scaler.fit_transform(y)
axis[1].plot(x, y_scal, color='orange', label='Normalized')
plt.legend()
plt.show()
```
